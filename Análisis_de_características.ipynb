```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
%matplotlib inline
```

```python
data_seg_selec = data_seg[['Mean', 'Median', 'Minimum', 'Maximum',
       'Variance', 'Skewness', 'Kurtosis', 'Energy', 'Entropy', 'IDM',
       'Contrast', 'Correlation', 'Type']]
data_seg_selec.head()
```

```python
features = ['Mean', 'Median', 'Minimum', 'Maximum',
       'Variance', 'Skewness', 'Kurtosis', 'Energy', 'Entropy', 'IDM',
       'Contrast', 'Correlation']
x = data_seg_selec.loc[:, features].values
y = data_seg_selec.loc[:,['Type']].values
```

```python
x = StandardScaler().fit_transform(x)  #normaliza las caracteristicas
np.mean(x),np.std(x)
```

```python
x = pd.DataFrame(data = x, columns = features)
x.head()
```



---



# PCA 2 componentes

```python
pca = PCA(n_components=2)

principalComponents = pca.fit_transform(x)
principalDf = pd.DataFrame(data = principalComponents
             , columns = ['Componente Principal 1', 'Componente Principal 2'])
principalDf.head()
```

```python
finalDf = pd.concat([principalDf, data_seg_selec[['Type']]], axis = 1)
finalDf.head()
```

```python
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Componente Principal 1', fontsize = 15)
ax.set_ylabel('Componente Principal 2', fontsize = 15)
ax.set_title('Componentes PCA', fontsize = 20)


Tipos = ['Normal', 'Grasa']
colors = ['b', 'y']
for Type, color in zip(Tipos,colors):
    indicesToKeep = finalDf['Type'] == Type
    ax.scatter(finalDf.loc[indicesToKeep, 'Componente Principal 1']
               , finalDf.loc[indicesToKeep, 'Componente Principal 2']
               , c = color
               , s = 75
               , alpha=0.7
               , linewidths = 0.8)
ax.legend(Tipos)
plt.savefig('PCA-2D.svg', format='svg', dpi=30)
```

```python
print(pca.explained_variance_ratio_)
```

```python
import seaborn as sns
df = pd.DataFrame({'Varianza Explicada':pca.explained_variance_ratio_,
             'Componentes Principales':['CP1','CP2']})
plt.figure(figsize=(4,4))
sns.set()
ax = sns.barplot(x='Componentes Principales',y="Varianza Explicada", 
           data=df, color="c")
ax.set_xlabel('Componentes Principales',fontsize=14)
ax.set_ylabel('Varianza Explicada',fontsize=14)
plt.savefig('Variance2D.svg', format='svg', dpi=30)
```

```python
finalDf['Type'] = finalDf['Type'].apply(lambda x: x.replace('Normal','0'))
finalDf['Type'] = finalDf['Type'].apply(lambda x: x.replace('Grasa','1'))
```

```python
finalDf['Type'].value_counts()
```

```python
finalDf.head()
finalDf.to_csv("/content/PCA2.csv", index=False, encoding='utf-8-sig')
```

# PCA 3 componentes

```python
pca3 = PCA(n_components=3)

principalComponents = pca3.fit_transform(x)
principalDf3 = pd.DataFrame(data = principalComponents
             , columns = ['Componente Principal 1', 'Componente Principal 2', 'Componente Principal 3'])
principalDf3.head()
```

```python
finalDf3 = pd.concat([principalDf3, data_seg_selec[['Type']]], axis = 1)
finalDf3.head(5)
```

```python
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(111, projection='3d') 
ax.set_xlabel('Componente Principal 1', fontsize = 15)
ax.set_ylabel('Componente Principal 2', fontsize = 15)
ax.set_zlabel('Componente Principal 3', fontsize = 15)
ax.set_title('Componentes PCA', fontsize = 20)

Tipos = ['Normal', 'Grasa']
colors = ['b', 'y']
for Type, color in zip(Tipos,colors):
    indicesToKeep = finalDf3['Type'] == Type
    ax.scatter(finalDf3.loc[indicesToKeep, 'Componente Principal 1']
               , finalDf3.loc[indicesToKeep, 'Componente Principal 2']
               , finalDf3.loc[indicesToKeep, 'Componente Principal 3']
               , c = color
               , s = 75
               , alpha=0.6
               , linewidths = 0.8)
ax.legend(Tipos)
plt.savefig('PCA-3D.svg', format='svg', dpi=30)
```

```python
pca3.explained_variance_ratio_
```

```python
import seaborn as sns
df = pd.DataFrame({'Varianza Explicada':pca3.explained_variance_ratio_,
             'Componentes Principales':['CP1','CP2','CP3']})

plt.figure(figsize=(4,4))
sns.set()
ax = sns.barplot(x='Componentes Principales',y="Varianza Explicada", 
           data=df, color="c")
ax.set_xlabel('Componentes Principales',fontsize=14)
ax.set_ylabel('Varianza Explicada',fontsize=14)
plt.savefig('Variance3D.svg', format='svg', dpi=30)
```

```python
finalDf3['Type'] = finalDf3['Type'].apply(lambda x: x.replace('Normal','0'))
finalDf3['Type'] = finalDf3['Type'].apply(lambda x: x.replace('Grasa','1'))
```
